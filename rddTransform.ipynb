```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("transformTut").getOrCreate()
```

```python
data = [("Alice", 23),("Bob", 35),("Charlie", 43),("Alice", 63)]
rdd = spark.sparkContext.parallelize(data)
```

```python
rdd.collect()
```

# Map operation

```python
mapped_rdd = rdd.map(lambda x: (x[0].upper(), x[1]))
```

```python
mapped_rdd.collect()
```

# Filter operation

```python
filtered_rdd = rdd.filter(lambda x: x[1] > 25)
```

```python
filtered_rdd.collect()
```

# ReduceByKey

```python
reduced_rdd = rdd.reduceByKey(lambda x, y: x + y)
```

```python
reduced_rdd.collect()
```

#### it keeps reducing like

(((v1 + v2) + v3) + v4) ...

#### another example

```python
rdd2 = spark.sparkContext.parallelize([
    ("a", "hello"),
    ("a", "world"),
    ("b", "spark"),
    ("a", "2")
])

result = rdd2.reduceByKey(lambda x, y: x + " " + y)
```

```python
result.collect()
```

# Sort

```python
sorted_rdd = rdd.sortBy(lambda x: x[1], ascending=True)
```

```python
sorted_rdd.collect()
```

# Save/Read RDDs to/from text files

```python
sorted_rdd.saveAsTextFile("output")  #saves each element to each txt file in "output" folder it creates
```

```python
rdd_text = spark.sparkContext.textFile("output") #reads from that folder
```

```python
rdd_text.collect()
```

```python
rdd_text1 = spark.sparkContext.textFile("data1.txt") #reads from that file but it parse it as raw string
# rdd_text1.first()
rdd_text1.collect()
```
